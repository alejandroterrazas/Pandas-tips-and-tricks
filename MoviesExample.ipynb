## Load two datasets-1) ratings and 2) movie names

```python
df_ratings = pd.read_csv('../Desktop/ml-20m/ratings.csv', header=0)
df_ratings.head()

df_movies = pd.read_csv('../Desktop/ml-20m/movies.csv', header=0)
```

```python
##extract the year from the movie title
def extract_year(name_string):
    #test = x.split("(")[1]
    return name_string[-5:-1]


#.str.split("(", n=1, expand=True)[1]

df_movies['year'] = df_movies['title'].apply(extract_year)
```

```python
def extract_genre_type(genre_string, genre_type):
    return True if genre_type in genre_string else False

df_movies['Adventure'] = df_movies['genres'].apply(extract_genre_type, 
                                                  genre_type='Adventure')

df_movies['Comedy'] = df_movies['genres'].apply(extract_genre_type, 
                                                  genre_type='Comedy')
df_movies.head()
#print(extract_movie_type('Adventure|Comedy', movie_type='Comedy'))
#df_movies['adventur']
```

```python
import seaborn as sns
```

```python
def number_genres(genre_string):
    return genre_string.count('|')+1
    
#number_genres('Adventure|Animation|Children|Comedy|Fantasy')
df_movies['n_genres'] = df_movies['genres'].apply(number_genres)
```

```python
df_movies['genre_ranking'] = df_movies.genres.rank(ascending=False)
df_movies.sort_values(['genre_ranking'])

#only keep the top .4
qvalue = df_movies.genre_ranking.quantile(.2)
df_movies = df_movies.loc[df_movies['genre_ranking'] > qvalue,:]
                        
```
